```python
from math import sqrt
def compute_distanse(hand):
    x = []
    y = []
    width = []
    for k in hand:
        x.append(k[0])
        y.append(k[1])
    for i in range(1, 21):
        width.append(sqrt((x[i] - x[0]) ** 2 + (y[i] - y[0]) ** 2))
    width.append(sqrt((x[8] - x[4]) ** 2 + (y[8] - y[4]) ** 2))
    width.append(sqrt((x[12] - x[8]) ** 2 + (y[12] - y[8]) ** 2))
    width.append(sqrt((x[16] - x[12]) ** 2 + (y[16] - y[12]) ** 2))
    width.append(sqrt((x[20] - x[16]) ** 2 + (y[20] - y[16]) ** 2))
    width = np.array(width)
    return width.reshape(1, 24)
```

```python
def s_df(gesture, gesture_num):
    indexes = ['label']
    for idx in range(0, 20):
        indexes.append('pos_'+str(idx))
    df = pd.DataFrame(columns=indexes)
    for i in range(1, 501):   
        arrayL = np.load(gesture+'L'+str(i)+'.npy')    
        distanseL = compute_distanse(arrayL)[0]
        d = [gesture_num]
        for idx in range(0, 20):
            d.append(distanseL[idx])
        dfL = pd.DataFrame([d], columns=indexes)
        df = df.append(dfL, ignore_index=True)

        arrayR = np.load(gesture+'R'+str(i)+'.npy')    
        distanseR = compute_distanse(arrayR)[0]
        d = [gesture_num]
        #print(d)
        for idx in range(0, 20):
            d.append(distanseL[idx])
        dfR = pd.DataFrame([d], columns=indexes)
        df = df.append(dfR, ignore_index=True)

    return df
```
